```python
"""
from Bio import SeqIO
from Bio.Seq import Seq

# Caminho para o arquivo FASTA
fasta_file = "/content/drive/MyDrive/AlgBio/dengue_virus_completo.fasta"

# Leitura do arquivo FASTA
for seq_record in SeqIO.parse(fasta_file, "fasta"):
    # Obtendo o identificador da sequência
    seq_id = seq_record.id
    # Obtendo a sequência
    sequence = str(seq_record.seq)

    print("ID da Sequência:", seq_id)
    print("Sequência:", sequence)

dna_seq = Seq(sequence)

protein_seq = dna_seq.translate()

print("Sequência de proteína traduzida:", protein_seq)

# Letra inicial e final do trecho desejado
letra_inicial = "M"  # Letra inicial do trecho
letra_final = "*"    # Letra final do trecho

# Lista para armazenar os trechos encontrados
trechos_encontrados = []

# Procurando os trechos
pos_inicial = None
pos_final = None
for i in range(len(protein_seq)):
    if protein_seq[i] == letra_inicial:
        pos_inicial = i
    elif protein_seq[i] == letra_final:
        if pos_inicial is not None:
            trecho_encontrado = protein_seq[pos_inicial:i+1]
            trechos_encontrados.append((trecho_encontrado, pos_inicial, i))
            pos_inicial = None

# Imprimindo os trechos encontrados
if trechos_encontrados:
    print("Trechos encontrados:")
    for trecho, pos_inicial, pos_final in trechos_encontrados:
        print("Trecho:", trecho)
        print("Posição inicial:", pos_inicial)
        print("Posição final:", pos_final)
        print()
else:
    print("Nenhum trecho encontrado.")

def traduzir_frames(dna_seq):
  sequencia = Seq(dna_seq)
  for frame in range(3):

    print(f"Frame {frame + 1}: {sequencia[frame:].translate()}")

traduzir_frames(dna_seq)
"""
"""
from Bio import SeqIO
from Bio.Seq import Seq

# Caminho para o arquivo FASTA
fasta_file = "/content/drive/MyDrive/AlgBio/dengue_virus_completo.fasta"

# Leitura do arquivo FASTA
for seq_record in SeqIO.parse(fasta_file, "fasta"):
    # Obtendo o identificador da sequência
    seq_id = seq_record.id
    # Obtendo a sequência
    sequence = str(seq_record.seq)

    print("ID da Sequência:", seq_id)
    print("Sequência:", sequence)

    # Traduzindo os três frames
    for frame in range(3):
        dna_seq = Seq(sequence[frame:])
        protein_seq = dna_seq.translate()

        print(f"Sequência de proteína traduzida para o frame {frame + 1}:", protein_seq)

        # Letra inicial e final do trecho desejado
        letra_inicial = "M"  # Letra inicial do trecho
        letra_final = "*"    # Letra final do trecho

        # Lista para armazenar os trechos encontrados
        trechos_encontrados = []

        # Procurando os trechos
        pos_inicial = None
        pos_final = None
        for i in range(len(protein_seq)):
            if protein_seq[i] == letra_inicial:
                pos_inicial = i
            elif protein_seq[i] == letra_final:
                if pos_inicial is not None:
                    trecho_encontrado = protein_seq[pos_inicial:i+1]
                    trechos_encontrados.append((trecho_encontrado, pos_inicial, i))
                    pos_inicial = None

        # Imprimindo os trechos encontrados
        if trechos_encontrados:
            print(f"Trechos encontrados para o frame {frame + 1}:")
            for trecho, pos_inicial, pos_final in trechos_encontrados:
                print("Trecho:", trecho)
                print("Posição inicial:", pos_inicial)
                print("Posição final:", pos_final)
                print()
        else:
            print(f"Nenhum trecho encontrado para o frame {frame + 1}.")
"""

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# Caminho para o arquivo FASTA
fasta_file = "/content/drive/MyDrive/AlgBio/dengue_virus_completo.fasta"

# Lista para armazenar os registros SeqRecord
records = []

# Leitura do arquivo FASTA
for seq_record in SeqIO.parse(fasta_file, "fasta"):
    # Obtendo o identificador da sequência
    seq_id = seq_record.id
    # Obtendo a sequência
    sequence = str(seq_record.seq)

    # Traduzindo os três frames
    for frame in range(3):
        dna_seq = Seq(sequence[frame:])
        protein_seq = dna_seq.translate()

        # Letra inicial e final do trecho desejado
        letra_inicial = "M"  # Letra inicial do trecho
        letra_final = "*"    # Letra final do trecho

        # Procurando os trechos
        pos_inicial = None
        pos_final = None
        protein_num = 1
        for i in range(len(protein_seq)):
            if protein_seq[i] == letra_inicial:
                pos_inicial = i
            elif protein_seq[i] == letra_final:
                if pos_inicial is not None:
                    trecho_encontrado = protein_seq[pos_inicial:i+1]
                    # Criando um novo SeqRecord para o trecho encontrado
                    record = SeqRecord(trecho_encontrado, id=f"{seq_id}, Frame {frame + 1}, proteína {protein_num}", description="")
                    # Adicionando o SeqRecord à lista
                    records.append(record)
                    protein_num += 1
                    pos_inicial = None

# Escrevendo os registros SeqRecord em um novo arquivo FASTA
SeqIO.write(records, "saida_possiveis_trechos.fasta", "fasta")
```
